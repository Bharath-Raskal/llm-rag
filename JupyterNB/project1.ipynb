**Objective:** Build a tool that takes a URL and provides a concise, intelligent summary of its content.

```python
import os

from dotenv import load_dotenv

load_dotenv()


def validate_environment_setup() -> None:
    """
    Verifies that all required API keys are properly configured in the environment.
    """
    key_configs = {
        "OPENAI_API_KEY": "sk-",
        "ANTHROPIC_API_KEY": "sk-ant-",
        "GEMINI_API_KEY": "AIza",
        "DEEPSEEK_API_KEY": "sk-",
    }

    print("--- Environment Configuration Check ---")
    for key, expected_prefix in key_configs.items():
        print(key)
        print(expected_prefix)
        value = os.getenv(key)
        print(value)
        if not value:
            print(f"❌ {key:<20}: Not found in .env")
            continue

        if not value.startswith(expected_prefix):
            print(f"⚠️ {key:<20}: Found, but expected prefix '{expected_prefix}'")
        elif value.strip() != value:
            print(f"⚠️ {key:<20}: Found, but contains hidden whitespace")
        else:
            print(f"✅ {key:<20}: Validated")


validate_environment_setup()
```
