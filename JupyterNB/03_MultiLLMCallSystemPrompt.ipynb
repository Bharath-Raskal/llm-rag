## M4 Quick Preview: Your First API Call

Test frontier model APIs using the OpenAI Python client.

```python
import os

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
```

```python
def quick_test_frontier_models() -> None:
    messages = [
        {"role": "system", "content": "You are a grumpy 5-year-old."},
        {"role": "user", "content": "What is the capital of the moon?"},
    ]

    # 1. Test OpenAI
    print("--- Testing OpenAI ---")
    openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY", "").strip())
    response = openai_client.chat.completions.create(
        model="gpt-5-mini",
        messages=messages,
    )
    print(response.choices[0].message.content)

    # 2. Test Deepseek (OpenAI-compatible API)
    print("\n--- Testing Deepseek ---")
    deepseek_client = OpenAI(
        base_url="https://api.deepseek.com/v1",
        api_key=os.getenv("DEEPSEEK_API_KEY", "").strip(),
    )
    response = deepseek_client.chat.completions.create(
        model="deepseek-chat",
        messages=messages,
    )
    print(response.choices[0].message.content)
```

```python
quick_test_frontier_models()
```
