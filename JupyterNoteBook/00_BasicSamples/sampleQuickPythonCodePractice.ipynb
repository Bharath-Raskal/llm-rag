```python
l =[1,2,3,4,4,5,6,6,7]
print(l)
```

```python
distinct =[]
duplicate=[]
for i in l :
    #print(i in distinct)
    if ((i in distinct)==True):
      #print('entered if')
      duplicate.append(i)  
    else:
       #print('entered else')
       distinct.append(i)
print(distinct)
print(duplicate)
```
