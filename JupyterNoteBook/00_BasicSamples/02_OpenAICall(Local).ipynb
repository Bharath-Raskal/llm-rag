# OpenAI Chat Completions

Connect to the OpenAI API using `OPENAI_API_KEY` from `.env` and fetch a chat completion response.

This notebook uses **`gpt-4o-mini`** — OpenAI's lowest-cost chat model, ideal for learning and experiments.

```python
import os

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY", "").strip()
if not api_key:
    raise ValueError("OPENAI_API_KEY not found. Add it to your .env file.")

client = OpenAI(api_key=api_key)
print("OpenAI client ready.")
```

```python
MODEL = "gpt-4o-mini"

response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": "Say hello and explain what a RAG pipeline does in one sentence."},
    ],
)
print(response)
#assistant_message = response.choices[0].message.content
#print(assistant_message)
```

```python
print(f"Model: {response.model}")
print(f"Finish reason: {response.choices[0].finish_reason}")
print(f"Prompt tokens: {response.usage.prompt_tokens}")
print(f"Completion tokens: {response.usage.completion_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")
```
