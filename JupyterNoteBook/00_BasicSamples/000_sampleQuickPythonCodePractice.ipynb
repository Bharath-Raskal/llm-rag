## 1.Duplicates and Distict List Separation Hard coding


```python
l =[1,2,3,4,4,5,6,6,7]
print(l)
distinct =[]
duplicate=[]
for i in l :
    #print(i in distinct)
    if ((i in distinct)==True):
      #print('entered if')
      duplicate.append(i)  
    else:
       #print('entered else')
       distinct.append(i)
print(distinct)
print(duplicate)



print('--------Using Set------') 
print(set(l))
for i in set(l):
   if(l.count(i)>1):
      print(i)
print('--------Using Set single like------') 
print([i for i in set(l) if l.count(i)>1])
```

## 2.Reverse String Sample

```python
#s = 'Mom'
s='ABCDEFG'
print(reversed(s))#returns reversed object need to convert to text back as below
r=''.join(reversed(s))
print(r)


print(s[2::])#initial slice
print(s[:2:])#end slice
print(s[::2])# slice step (size) can also be used for reversing
print(s[::-1])
print(s[:-1])#slices last letter

print ('--------------')
r=''
for i in range(len(s)-1,-1,-1):
    r+=s[i]
print(r)
```

## palindrome

```python
from numpy.char import isalnum


s='Racecar '
s=''.join (clean_s for clean_s in s if clean_s.isalnum())#isalnum() checks for specialcharacter
s_lower=s.lower()
print(s_lower)

if (s_lower[::-1]==s_lower):
    print(s,'is a palindrome')
else:
    print(s,'is not a palindrome')
```

##Random


```python
import random

def randomNumbergame():
    play_again = 'y'
    while play_again.lower() == 'y':
        x = random.randint(1,100)
        guesscount = 0
        while True:
            try:
                user_input = int(input("Guess the Number : "))
            except ValueError:
                print("Invalid input! Please enter a number.")
                continue
            guesscount += 1
            if user_input == x:
                print(f"You guessed Correctly, Number of Attempts : {guesscount}")
                break
            elif user_input < x:
                print(f"{user_input} is too low! Try again.")
            elif user_input > x:
                print(f"{user_input} is too high! Try again.")
        while True:
            play_again = input("Do you want to retry Y/N :").lower()
            if play_again in ['y', 'n']:
                break
            print("Invalid input! Please enter Y or N.")
    if play_again == 'n':
        print("Thanks for playing!")

randomNumbergame()

```
