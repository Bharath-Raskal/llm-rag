### <span style="color:lightgreen"> PreProcess Sequential Data </span>

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Step 1: Create a Sample Dataset
data = {
    "Date": pd.date_range(start="2025-01-01", periods=10, freq="D"),
    "Temperature": [30, 32, None, 35, 36, None, 40, 41, 43, None]
}
df = pd.DataFrame(data)

# Display Original Data (With Missing Values)
print("\nOriginal Data (With Missing Values):")
print(df)

# Step 2: Handle Missing Values (Using Interpolation)
df["Temperature"] = df["Temperature"].interpolate()
# Display Interpolated Data
print("\nData After Interpolation:")
print(df)

# Step 3: Normalize Data Using Min-Max Scaling
# Min-Max Scaling scales data between 0 and 1 using
scaler = MinMaxScaler()
df["Temperature_Scaled"] = scaler.fit_transform(df["Temperature"].values.reshape(-1, 1))

# Display Normalized Data
print("\nData After Min-Max Normalization:")
print(df[["Date", "Temperature_Scaled"]])

# Step 4: Convert data into sequences for RNN input
sequence_length = 3  # Use past 3 days to predict the next day

def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

X, y = create_sequences(df["Temperature_Scaled"].values, sequence_length)

# Display Sequences
print("\nGenerated Sequences for RNN Input:")
for i in range(len(X)):
    print(f"X[{i}]: {X[i]}  -->  y[{i}]: {y[i]}")

# Step 5: Train-Test Split (80-20)
split_index = int(len(X) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"\nTrain size: {len(X_train)}, Test size: {len(X_test)}")

# Step 6: Text-Based Visualization using ASCII Art
print("\nTemperature Trends:")
for i, temp in enumerate(df["Temperature"]):
    bar = "█" * int(temp - min(df["Temperature"]))  # Create bars based on temperature
    print(f"{df['Date'][i]} | {temp:.1f}°C | {bar}")

```
