### <span style="color:lightgreen"> Basic RNN Application </span>

```python
# Import Statements
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# SimpleRNN - A Recurrent Neural Network (RNN) layer that processes sequential data. 
# It has a memory mechanism that helps capture patterns over time.

# Input Data - Sample temperature data (last 7 days)
temperature_data = [10, 11, 12, 13,14, 15]


# Data Preparation - Reshaping for RNN Input - [samples, time steps, features]
inputData = np.array(temperature_data[:]).reshape(1, 6, 1)
output = np.array(temperature_data[-1])
#print(inputData)
#print(output)

# Define RNN model
model = Sequential()
model.add(SimpleRNN(units=100, input_shape=(6,1)))
model.add(Dense(1))

# Compile model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(inputData, np.array([output]), epochs=1000, verbose=0)

# Make a prediction for the next day
temp_prediction = model.predict(inputData)
print(f"Predicted Temperature: {temp_prediction[0][0]:.2f} °C")
```
